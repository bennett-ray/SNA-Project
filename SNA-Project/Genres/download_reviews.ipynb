```python
import urllib.request, os, re, time

def connect_to_page(url):
    req = urllib.request.Request(url, headers={'User-Agent' : "Mozilla/5.0 (Windows NT 6.0; WOW64; rv:12.0) Gecko/20100101 Firefox/12.0"}) 
    page = urllib.request.urlopen(req).read().decode('utf-8')
    return(page)
```

```python
def make_file(content, name):
    path = os.path.join(os.getcwd(), 'Reviews', name)
    review_file = open(path, 'w', encoding='utf-8')
    review_file.write(content)
    review_file.close()

def url_crawler(url, checker, all_urls):
    print(url)
    page = re.sub('[\r\n\t]', '', connect_to_page(url))
    while re.search('  ', page) != None:
        page = re.sub('  ', ' ', page)
    
    reviews = re.findall('<span class="all-text">(.+?)</span>', page)
    urls = re.findall('<a class="rev-title" href=\"([^\"]+)\"', page)
    
    for index in range(len(reviews)):
        if urls[index] not in all_urls:
            review = re.sub('<br.+?>', ' ', reviews[index])
            name = re.search('turpravda\.ru/(.+?)\.html', urls[index])
            name = re.sub('[/-]', '_', name.group(1)) + '.txt'
            make_file(review, name)
        else:
            checker = False
            break
            
    if urls == []:
        checker = False
        
    all_urls += urls
    
    return(checker, all_urls)

def hotel_crawler(url):
    number = 1
    all_urls = []
    checker = True
    while checker == True:
        start_url = url+'?p='+str(number)+'#reviews'
        checker, all_urls = url_crawler(start_url, checker, all_urls)
        number += 1

def hotels_in_city(url, checker, all_hotels):
    print('hotels_in_city: ', url)
    page = re.sub('[\r\n\t]', '', connect_to_page(url))
    while re.search('  ', page) != None:
        page = re.sub('  ', ' ', page)
    hotels_urls = re.findall('<a class="hotel-name-title\" href=\"([^\"]+)\"', page)
    for index in range(len(hotels_urls)):
        if hotels_urls[index] not in all_hotels:
            hotel_url = 'https://www.turpravda.ru' + hotels_urls[index]
            hotel_crawler(hotel_url)
        else:
            checker = False
            break
    all_hotels += hotels_urls
    return(checker, all_hotels)
        
number = 1
checker = True
all_hotels = []
while checker == True:
    city_hotels = 'https://www.turpravda.ru/eg/taba/hotels.html?p='+str(number)
    checker, all_hotels = hotels_in_city(city_hotels, checker, all_hotels)
    number += 1
```
