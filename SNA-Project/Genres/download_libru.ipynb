```python
import urllib.request, os, re, time
from bs4 import BeautifulSoup

def connect_to_page(url):
    try:
        req = urllib.request.Request(url, headers={'User-Agent' : "Mozilla/5.0 (Windows NT 6.0; WOW64; rv:12.0) Gecko/20100101 Firefox/12.0"}) 
        page = urllib.request.urlopen(req).read().decode('cp1251')
    except urllib.error.HTTPError:
        page = ''
    except urllib.error.URLError:
        time.sleep(2)
        page = connect_to_page(url)

    return(page)
```

```python
def make_file(content, name):
    path = os.path.join(os.getcwd(), 'prose_classic_try', name)
    review_file = open(path, 'w', encoding='utf-8')
    review_file.write(content)
    review_file.close()

def pages_crawler(url, checker, all_urls, all_genres):
    page = connect_to_page(url)
    if page != '':
        all_urls += re.findall('</a>: <a href=([^>]+)>', page)
        all_genres += re.findall('(Оценка:.+?)?&nbsp;\s+([А-ЯЁа-яё ]+)(\s+<A HREF=[^>]+>Комментарии</A>)?\s+</small><br>', page)
    else:
        checker = False
    return(checker, all_urls, all_genres)

base_url = 'http://az.lib.ru/janr/index_janr_1-'
number = 1
checker = True
all_books, all_genres = [], []

while checker == True:
    print('Loading the '+str(number)+' page...')
    url = base_url+str(number)+'.shtml'
    checker, all_books, all_genres = pages_crawler(url, checker, all_books, all_genres)
    number += 1

for book_url_index in range(len(all_books)):
    print('http://az.lib.ru'+all_books[book_url_index])
    book_text = ''
    page = connect_to_page('http://az.lib.ru'+all_books[book_url_index])
    if page != '':
        page = re.sub('[\r\n\t]', ' ', page)
        page = re.sub('&nbsp;', '', page)
        while re.search('--', page) != None: page = re.sub('--', '-', page)
        while re.search('  ', page) != None: page = re.sub('  ', ' ', page)
        content = re.findall('<!\- Собственно произведение \->(.+?)<!\->', page)
        book_text = re.sub('<[^>]+>', '', content[0])
    
    genre = re.sub(' ', '_', all_genres[book_url_index][1])
    name = genre + '_' + re.search('/(.+?)\.shtml', all_books[book_url_index]).group(1) + '.txt'
    name = re.sub('/', '_', name)
    make_file(book_text, name)
    
```
