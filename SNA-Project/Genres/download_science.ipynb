```python
import urllib.request, os, re, time
from bs4 import BeautifulSoup

def connect_to_page(url):
    try:
        req = urllib.request.Request(url, headers={'User-Agent' : "Mozilla/5.0 (Windows NT 6.0; WOW64; rv:12.0) Gecko/20100101 Firefox/12.0"}) 
        page = urllib.request.urlopen(req).read().decode('utf-8')
    except TimeoutError:
        time.sleep(5)
        page = connect_to_page(url)
    return(page)
```

```python
def pages_crawler(url, checker, all_urls):
    page = connect_to_page(url)
    urls_on_page = re.findall('<h2><a href=\"([^\"]+)\"', page)
    if (all_urls != [] and urls_on_page[0] != all_urls[0]) or all_urls == []:
        all_urls += urls_on_page
    else:
        checker = False
    return(checker, all_urls)

def make_file(content, name):
    path = os.path.join(os.getcwd(), 'Science', name)
    review_file = open(path, 'w', encoding='utf-8')
    review_file.write(content)
    review_file.close()

base_url = 'https://www.nkj.ru/news/index.php?mobile=N&news_search_DATE_CREATE_1=&news_search_DATE_CREATE_1_DAYS_TO_BACK=85&news_search_DATE_CREATE_2=&set_filter=Y&PAGEN_2='
number = 1
checker = True
all_news = []
while checker == True:
    print('Loading the '+str(number)+' page...')
    url = base_url+str(number)
    checker, all_news = pages_crawler(url, checker, all_news)
    number += 1
    
for news in all_news:
    soup = BeautifulSoup(connect_to_page('https://www.nkj.ru'+news), 'html.parser')
    [x.extract() for x in soup.findAll('script')]
    [x.extract() for x in soup.findAll('div', id="article_slider")]
    main = soup.find('main')
    content_text = ''
    for el in main.findAll('p'):
        el = re.sub('[\t\r\n]+', '', str(el.text))
        content_text += el
        
    name = re.search('([0-9]+)', news)
    make_file(content_text, name.group(1)+'.txt')
    
```
